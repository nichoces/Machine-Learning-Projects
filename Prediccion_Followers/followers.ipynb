```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from autots import AutoTS
```

```python
url = "https://raw.githubusercontent.com/amankharwal/Website-data/master/stats.csv"

dataframe = pd.read_csv(url)
dataframe.head()
```

```python
dataframe.dtypes
```

```python
dataframe.shape
```

```python
plt.figure(figsize=(15, 10))
sns.set_theme(style="whitegrid")
plt.title("Seguidores mensuales")
sns.barplot(x="followers_gained", y="period_end", data=dataframe)
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.set_theme(style="whitegrid")
plt.title("Seguidores totales mensuales")
sns.barplot(x="followers_total", y="period_end", data=dataframe)
plt.show()
```

```python

model = AutoTS(forecast_length=4, frequency='infer', ensemble='simple')
model = model.fit(dataframe, date_col='period_end', value_col='followers_gained', id_col=None)
prediction = model.predict()
prediction
```
