```python
# Importamos librerías

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creamos el dataframe
url = r'C:\Users\34722\Documents\DataScience\EjerciciosResueltos\Entregas\Prediccion_Coches_Usados\data\car_price_data1.csv'
dataframe = pd.read_csv(url)

# Vemos el número de registros
dataframe.shape[0]

# Vemos lo que contiene el dataframe
dataframe.head()

# Vemos el tipo de las columnas

dataframe.dtypes

# Eliminamos las columnas que no vamos a usar de momento
dataframe= dataframe.drop('Model',axis=1)

# Miramos los nulos
dataframe.isna().sum()

# Eliminamos los nulos y comprobamos que no queden
dataframe = dataframe.dropna()
dataframe.isna().sum()

```

```python
# Pintamos la distribución de coches segun su precio

plt.figure(figsize=[11,5])
sns.distplot(dataframe['Price'])
plt.title('Cantidad de coches segun su precio')
```

```python
# Vemos los outliers según los cuartiles

# Eliminamos el 5 % más caros
q = dataframe['Price'].quantile(0.99)
dataframe = dataframe[dataframe['Price']<q]
# Eliminamos el 5 % con más kilometros
q = dataframe['Mileage'].quantile(0.99)
dataframe = dataframe[dataframe['Mileage']<q]
# Eliminamos el 5 % más antiguos
q = dataframe['Year'].quantile(0.01)
dataframe = dataframe[dataframe['Year']>q]

# Vemos el número de columnas que hemos eliminado
dataframe.shape[0]
```

```python
# Pintamos la distribución de coches segun su precio despues de eliminar outliers

plt.figure(figsize=[11,5])
sns.distplot(dataframe['Price'])
plt.title('Cantidad de coches segun su precio')
```
